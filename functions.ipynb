```python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
    
num1 = 10
num2 = 25
num3 = 15

print("The maximum number is:", find_max(num1, num2, num3))
```
