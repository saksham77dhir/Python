```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
df=pd.DataFrame({
    'x':list(range(-100,100,5))
})
df.head()
```

```python
plt.plot(df['x'])
```

```python
df['x2']=df['x']**2
df['x3']=df['x']**3
df['x4']=df['x']**4

df.head()
```

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(1,4,figsize=(20,5))
ax1.plot(df['x'],color='red',label='degree 1',marker='o')
ax2.plot(df['x2'],color='blue',label='degree 2',marker='o')
ax3.plot(df['x3'],color='green',label='degree 3',marker='o')
ax4.plot(df['x4'],color='black',label='degree 4',marker='o')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
plt.show()
```

```python
df=pd.read_csv('Position_Salaries.csv')
df
```

```python
plt.scatter(df['Level'],df['Salary'],color='lightcoral')
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
X=df[['Level']]
y=df[['Salary']]
```

```python
regressor=LinearRegression()
regressor.fit(X,y)
```

```python
pr=PolynomialFeatures(degree=5)
X_poly=pr.fit_transform(X)
PolyReg=LinearRegression()
PolyReg.fit(X_poly,y)
```

```python
y_pred_lr=regressor.predict(X)
y_pred_poly=PolyReg.predict(X_poly)
```

```python
plt.scatter(X,y,color='lightcoral')
plt.plot(X,PolyReg.predict(X_poly),color='firebrick')
plt.box(False)
plt.show()
```
