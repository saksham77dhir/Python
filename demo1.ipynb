```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# Widgets
name = widgets.Text(description="Name:")
emoji = widgets.Dropdown(
    options=["😊", "👋", "🎉", "🤖", "🔥"],
    description="Emoji:"
)
fontsize = widgets.IntSlider(value=16, min=10, max=50, step=2, description="Font Size:")
output = widgets.Output()

greet_button = widgets.Button(description="Greet", button_style="success")
clear_button = widgets.Button(description="Clear", button_style="danger")

# Greet button logic
def greet_user(b):
    with output:
        output.clear_output()
        style = f"font-size:{fontsize.value}px;"
        print_html = f"<p style='{style}'>Hello, {name.value}! {emoji.value}</p>"
        display(HTML(print_html))

# Clear button logic
def clear_output_area(b):
    output.clear_output()

# Bind events
greet_button.on_click(greet_user)
clear_button.on_click(clear_output_area)

# Display
display(widgets.VBox([
    name,
    emoji,
    fontsize,
    widgets.HBox([greet_button, clear_button]),
    output
]))
```
