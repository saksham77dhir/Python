```python
import ipywidgets as widgets
from IPython.display import display, clear_output

category_dropdown = widgets.Dropdown(options=['Length', 'Weight', 'Temperature'], description='Category:')

input_value = widgets.FloatText(value=1.0, description='Value:')
from_unit = widgets.Dropdown(description='From:')
to_unit = widgets.Dropdown(description='To:')
convert_button = widgets.Button(description='Convert', button_style='primary')
output = widgets.Output()

unit_options = {
    'Length': ['meter', 'kilometer', 'centimeter', 'inch', 'foot'],
    'Weight': ['kilogram', 'gram', 'pound', 'ounce'],
    'Temperature': ['celsius', 'fahrenheit', 'kelvin']
}

conversion_factors = {
    'meter': 1, 'kilometer': 1000, 'centimeter': 0.01, 'inch': 0.0254, 'foot': 0.3048,
    'kilogram': 1, 'gram': 0.001, 'pound': 0.453592, 'ounce': 0.0283495
}

def update_units(*args):
    from_unit.options = unit_options[category_dropdown.value]
    to_unit.options = unit_options[category_dropdown.value]

category_dropdown.observe(update_units, names='value')
update_units()

def convert_units(btn):
    with output:
        clear_output()
        val = input_value.value
        f = from_unit.value
        t = to_unit.value
        cat = category_dropdown.value

        if cat == 'Temperature':
            if f == t:
                result = val
            elif f == 'celsius':
                result = (val * 9/5 + 32) if t == 'fahrenheit' else (val + 273.15)
            elif f == 'fahrenheit':
                result = (val - 32) * 5/9 if t == 'celsius' else ((val - 32) * 5/9 + 273.15)
            elif f == 'kelvin':
                result = (val - 273.15) if t == 'celsius' else ((val - 273.15) * 9/5 + 32)
        else:
            result = val * conversion_factors[f] / conversion_factors[t]

        print(f"{val} {f} = {result:.4f} {t}")

convert_button.on_click(convert_units)

ui = widgets.VBox([
    widgets.HTML(value='<h3>🔄 Unit Converter</h3>'),
    category_dropdown,
    input_value,
    from_unit,
    to_unit,
    convert_button,
    output
])

ui

```
