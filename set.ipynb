```python
a={1,2,3,4}
b={5,6,7,8}
print(a)
print(b)
```

```python
#union
print(a.union(b))
print(a|b)
```

```python
#intersection
print(a.intersection(b))
print(a&b)
```

```python
print(a.difference(b))
print(b-a)
```

```python
#symmetric difference(removing common part)
print(a.symmetric_difference(b))
```
