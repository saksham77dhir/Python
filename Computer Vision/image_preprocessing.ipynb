```python
import cv2
```

```python
image=cv2.imread('download.jpeg')
```

```python
if image is not None:
    cv2.imshow('Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
image
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow('My image',gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurred_image=cv2.GaussianBlur(image,(15,15),0)
cv2.imshow('Blurred image',blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edges=cv2.Canny(gray_image,100,200)
cv2.imshow('Edges',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
small_image=cv2.resize(image,(600,400))
cv2.imshow('Small image',small_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image.shape
```

```python
cropped_image=image[25:300,150:900]
cv2.imshow('Cropped image',cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
drawing=image.copy()
cv2.rectangle(drawing,(50,50),(200,200),(36,200,150),2) #image,start_point,end_point,color,thickness
cv2.imshow('Rectangle',drawing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
