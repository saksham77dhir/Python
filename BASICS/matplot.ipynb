```python
import matplotlib.pyplot as plt
```

```python
#creating a figure and axis
fig, ax=plt.subplots(figsize=(10,6))
#setting background color
fig.set_facecolor('yellow')
ax.set_facecolor('lightblue')
#setting title and labels
ax.set_title('Sample plot',fontsize=20,color='blue')
ax.set_xlabel('X-axis label',fontsize=14,color='green')
ax.set_ylabel('Y-axis label',fontsize=14,color='green')
#plotting data
x=[1,2,3,4,5,6,7,8,9]
y=[2,3,5,7,40,11,30,25,23]
ax.plot(x,y,
        marker='o',
        markersize=10,
        markerfacecolor='red',
        markeredgecolor='black',
        markeredgewidth=2,
        color='black',
        linewidth=2,
        linestyle=':')
#line partition
ax.axhline(y=15,color='r')
ax.axvline(x=5,color='g')

#adding text
ax.text(5,16,'Text Annotation',fontsize=12,color='purple')
```
