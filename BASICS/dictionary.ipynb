```python
student={
    "name":"Alice",
    "age":20,
    "course":"btech",
    "marks":87
}
print(student)
```

```python
info=dict(name="alice",age=20,course="btech",marks=87)
print(info)
```

```python
temp={
    'lucknow':{30,35,40},
    'delhi':{35,40,45},
    "chennai":{20,25,30}
}
print(temp)
```

```python
#1st method to access value through key
info['name']
temp['delhi']
```

```python
#2nd method
info.get("name")
```

```python
#adding new key and value pair in dictionary
info['city']='lucknow'
print(info)
```
