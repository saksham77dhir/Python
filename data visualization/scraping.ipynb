```python
from dputils.scrape import Scraper, Tag
from pprint import pp
```

* Tag is a class which helps to decide what to collect from the page
* Scraper executes the scraping process

```python
url="https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=b9d06eec-2623-4e62-bf4d-63eb2599b4a3"
```

```python
scraper=Scraper(url)
```

```python
result=scraper.get_repeating_date_from_page(
    target=Tag('div',cls='DOjaWF gdgoEp'),
    items=Tag('div',cls='_75nlfW'),
    title=Tag('div',cls='KzDlHZ'),
    price=Tag('div',cls='Nx9bqj _4b5DiR')
)
```
